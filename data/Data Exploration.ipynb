# AppleStore Dataset Analysis

```python
import pandas as pd

appstore_df = pd.read_csv('AppleStore.csv', index_col=0)
```

## Dataset Understanding

```python
appstore_df.head()
```

```python
appstore_df.info()
```

```python
appstore_df['prime_genre'].unique()
```

## Extract most rated News application

```python
news_df = appstore_df[appstore_df['prime_genre'] == 'News']
news_df = news_df.sort_values(by='rating_count_tot', ascending=False)
most_rated_news_app = news_df.iloc[0]
```

```python
print('{} is the most rated News application'.format(most_rated_news_app['track_name']))
```

## Top 10 - Music Apps

```python
musics_df = appstore_df[appstore_df['prime_genre'] == 'Music']
musics_df = musics_df.sort_values(by='rating_count_tot', ascending=False)
top10_music_apps = musics_df.iloc[:10]
```

```python
top10_music_apps.head(10)
```

## Top 10 - Book Apps

```python
books_df = appstore_df[appstore_df['prime_genre'] == 'Book']
books_df = books_df.sort_values(by='rating_count_tot', ascending=False)
top10_books_apps = books_df.iloc[:10]
```
